```python
# Load .csv data
df = pd.read_csv("./data/Zip_Zhvi_AllHomes.csv", engine='python')
```

```python
# Find Philadelphia ZIP code with the largest average annual ZHIV value for each year

# Load the ZHVI data for each ZIP code, selecting only Philadelphia ZIP codes.
# Calculate the annual average ZHVI for each ZIP code in Philadelphia and each year.
# Identify the ZIP code with the maximum value for each year.
# The final result should be the year and the ZIP code with the maximum value.
```

```python
df.head() #check for correct data upload
```

```python
### Load the ZHVI data for each ZIP code, selecting only Philadelphia ZIP codes.

# Select Pennsylvania only
pennsylvania = df.loc[df['State']=='PA']
pennsylvania.head() #check for correct selection

# See number of columns and rows
pennsylvania.shape #(878,279)

# Select Philadelphia only
philly = pennsylvania.loc[df['City']=='Philadelphia']
philly.head() #check for correct selection

# See number of columns and rows
philly.shape #(36,279)
```

```python
philly.head()
```

```python
# Calculate the annual average ZHVI for each ZIP code in Philadelphia and each year.

#Select columns of interest
phillySelected = philly.iloc[:, np.r_[1:4,7:279]]
    # Question: 
    # When "philly.iloc[:, np.r_[1:4,7:279]]" 1:4 means columns 1(inclusive) through 4(exclusive), hence columns 1-3;
    # When calling "philly.iloc[:, np.r_[1:4,7:]]" the 7: call selects all random columns;
    # When calling "philly.iloc[:, np.r_[1:4,7:280]]" gives error since there is no column 280;
    # But calling "philly.iloc[:, np.r_[1:4,7:279]]" selects the rest of the columns (till end) and not until col 278;
phillySelected.head()
```

```python
# Transform dataframe from wide to long format
ZHVI = pd.melt(phillySelected, id_vars=['RegionName','City','State'], 
                value_name='ZHVI', var_name='Date')
ZHVI.head()
```

```python
ZHVI.shape
```

```python
# Convert Date to data
ZHVI['Date'] = pd.to_datetime(ZHVI['Date'])
ZHVI.head()
```

```python
# Rename RegionName to ZIP
ZHVI.columns = ['ZIP','City','State','Date','ZHVI']

# Separate by Month and Year columns
ZHVI['Month'] = ZHVI['Date'].dt.month
ZHVI['Year'] = ZHVI['Date'].dt.year

ZHVI.head()
```

```python
ZHVI.shape
```

```python
### Calculate the annual average ZHVI for each ZIP code in Philadelphia and each year.

annual_ZHVI = ZHVI.groupby(['ZIP', 'Year'])['ZHVI'].mean() 
```

```python
print(type(annual_ZHVI))
```

```python
annual_ZHVI.head()
```

```python
annual_ZHVI = annual_ZHVI.reset_index()
annual_ZHVI.head()
```

```python
annual_ZHVI.shape
```

```python
# Identify the ZIP code with the maximum value for each year.

yearMaxZIP = annual_ZHVI.loc[annual_ZHVI.groupby('Year')['ZHVI'].idxmax()]
    #https://stackoverflow.com/questions/32459325/python-pandas-dataframe-select-row-by-max-value-in-group
```

```python
# The final result should be the year and the ZIP code with the maximum value.
yearMaxZIP
```
